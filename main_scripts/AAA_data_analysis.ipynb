```python
import pandas as pd

# Load the HDF5 file
file_path = "geo_multigrid/results/SIMPLE_Re100_mesh127x127_profile.h5"

# Open the store and display the keys
with pd.HDFStore(file_path, 'r') as store:
    print(f"Keys in the HDF5 file: {store.keys()}")
    
    # Display each dataset
    for key in store.keys():
        print(f"\n=== {key} ===")
        print(store[key])
```
